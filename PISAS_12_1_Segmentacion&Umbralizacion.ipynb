## T12: Segmentación 
La segmentación subdivide una imagen en sus regiones constituyentes u objetos.  El nivel en el que la segmentación debe detenerse depende de los objetos de interés para la aplicación que deban ser aislados. En general las técnicas de segmentación se basan en la semejanza o la disncontinuidad de alguna propiedad, muy frecuentemente el valor del nivel digital. Por ejemplo la detección de perfiles (edges), puntos o líneas son un ejemplo de técnicas basadas en la discontinuidad. Mientras que la umbralización o la agregación  lo son de las técnicas basadas en la semejanza de los Niveles Digitales. 
Las técnicas de umbralización son unas de las mas populares por su rapidez. Veamos algunas de esas técnicas a continuación.

### 12.1 Umbralización
**cv2.threshold(src, thresh, maxval, type[, dst]) → retval, dst.**

    src – input array (single-channel, 8-bit or 32-bit floating point).
    dst – output array of the same size and type as src.
    thresh – threshold value.
    maxval – maximum value to use with the THRESH_BINARY and THRESH_BINARY_INV thresholding types.
    type – thresholding type (see the details below).
    
    Los métodos de umbralizacion que incluye esta funciín de OpenCV son lo siguientes:
* cv2.THRESH_BINARY
* cv2.THRESH_BINARY_INVY
* cv2.THRESH_TRUNCY
* cv2.THRESH_TOZEROY
* cv2.THRESH_TOZERO_INVY

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('gradient.png',0)
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)

titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in np.arange(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
